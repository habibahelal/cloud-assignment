```python
df.isnull().sum()
#df = df.dropna()
```

```python
df = df.dropna()
df.drop_duplicates()
```

```python
# Filter harry potter books
harry_potter_books = df[df['original_title'].str.contains("Harry Potter", case=False, na=False)]
harry_potter_books
```

```python
# Find the most selling books
most_sold_books = harry_potter_books.sort_values(by='books_count', ascending=False)
print(most_sold_books[['original_title', 'books_count']])


```

```python
# Calculate the average rating
average_rating = harry_potter_books['ratings_count'].mean()
print("Average Rating for Harry Potter Books:", average_rating)
```
